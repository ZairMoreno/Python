```python
# actividad 07 Python Zair Orlando Moreno Camarillo

import random

# Validar número entero
def valida_entero(mensaje, min_val, max_val):
    while True:
        try:
            num = int(input(mensaje))
            if min_val <= num <= max_val:
                return num
            else:
                print(f"Por favor, ingresa un número entre {min_val} y {max_val}.")
        except ValueError:
            print("Error: Ingresa un número válido.")

# Ejercicio 1
def funcion1():
    mascotas = ["SCARLETT", "GATINA", "PELUSA", "PELUSINA", "MILO", "KHALIFA"]
    for i, mascota in enumerate(mascotas):
        print(f"[{i}] {mascota} --> {len(mascota)} CARACTERES")

# Ejercicio 2
def generar_lista_aleatoria():
    lista = random.sample(range(30, 51), 10)
    return lista

def imprimir_lista(lista):
    for i, valor in enumerate(lista):
        print(f"[{i}] {valor}")

def funcion2():
    lista = generar_lista_aleatoria()
    imprimir_lista(lista)

# Ejercicio 3
def suma_listas(lista1, lista2):
    if len(lista1) != len(lista2):
        print("Las listas no son del mismo tamaño. Se usará el tamaño de la lista más pequeña.")
    tamaño = min(len(lista1), len(lista2))
    suma = [lista1[i] + lista2[i] for i in range(tamaño)]
    return suma

def funcion3():
    lista1 = [valida_entero(f"Ingrese el valor {i+1} para la primera lista: ", -100, 100) for i in range(5)]
    lista2 = [valida_entero(f"Ingrese el valor {i+1} para la segunda lista: ", -100, 100) for i in range(5)]
    resultado = suma_listas(lista1, lista2)
    print(f"Lista resultante: {resultado}")

# Ejercicio 4
def eliminar_duplicados(lista):
    return list(set(lista))

def funcion4():
    lista = []
    cantidad = valida_entero("¿Cuántos números deseas ingresar en la lista?: ", 1, 20)
    for i in range(cantidad):
        lista.append(valida_entero(f"Ingrese el número {i+1}: ", -100, 100))
    print(f"Lista original: {lista}")
    lista_sin_duplicados = eliminar_duplicados(lista)
    print(f"Lista sin duplicados: {lista_sin_duplicados}")

# Ejercicio 5
def calcular_media_mediana(lista):
    lista.sort()
    media = sum(lista) / len(lista)
    mediana = 0
    n = len(lista)
    if n % 2 == 0:
        mediana = (lista[n//2 - 1] + lista[n//2]) / 2
    else:
        mediana = lista[n//2]
    return media, mediana

def funcion5():
    lista = []
    cantidad = valida_entero("¿Cuántos números deseas ingresar en la lista?: ", 1, 20)
    for i in range(cantidad):
        lista.append(valida_entero(f"Ingrese el número {i+1}: ", -100, 100))
    media, mediana = calcular_media_mediana(lista)
    print(f"Lista: {lista}")
    print(f"Media: {media}, Mediana: {mediana}")


def menu():
    while True:
        print("\nSeleccione la función que desea ejecutar:")
        print("1. Imprimir nombres de mascotas y cantidad de caracteres")
        print("2. Generar y mostrar lista de números aleatorios")
        print("3. Sumar elementos de dos listas")
        print("4. Eliminar duplicados de una lista")
        print("5. Calcular media y mediana de una lista")
        print("6. Salir")

        opcion = valida_entero("Ingrese el número de su elección: ", 1, 6)

        if opcion == 1:
            funcion1()
        elif opcion == 2:
            funcion2()
        elif opcion == 3:
            funcion3()
        elif opcion == 4:
            funcion4()
        elif opcion == 5:
            funcion5()
        elif opcion == 6:
            print("Saliendo del programa...")
            break


menu()
```
