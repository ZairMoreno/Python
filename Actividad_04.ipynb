```python
import random
import math

def calcular_promedio():
    try:
        calificaciones = []
        for i in range(3):
            calificacion = float(input(f"Ingresa la calificación {i+1}: "))
            calificaciones.append(calificacion)

        promedio = sum(calificaciones) / len(calificaciones)

        if promedio < 0 or promedio > 100:
            print("Error en promedio")
        elif promedio < 30:
            print("Repetir")
        elif promedio < 60:
            print("Extraordinario")
        elif promedio < 70:
            print("Suficiente")
        elif promedio < 80:
            print("Regular")
        elif promedio < 90:
            print("Bien")
        elif promedio < 98:
            print("Muy bien")
        else:
            print("Excelente")
    except ValueError:
        print("Error: Ingrese un número válido.")

def calcular_salario():
    try:
        horas_trabajadas = float(input("Ingresa las horas trabajadas en la semana: "))
        salario_por_hora = float(input("Ingresa el salario por hora: "))

        if horas_trabajadas <= 40:
            salario_normal = horas_trabajadas * salario_por_hora
            salario_extra = 0
        elif horas_trabajadas <= 49:
            salario_normal = 40 * salario_por_hora
            salario_extra = (horas_trabajadas - 40) * salario_por_hora * 2
        else:
            salario_normal = 40 * salario_por_hora
            salario_extra = (9 * salario_por_hora * 2) + ((horas_trabajadas - 49) * salario_por_hora * 3)

        salario_total = salario_normal + salario_extra

        print(f"Salario por hora: {salario_por_hora}")
        print(f"Horas trabajadas: {horas_trabajadas}")
        print(f"Salario normal: {salario_normal}")
        print(f"Salario extra: {salario_extra}")
        print(f"Salario total: {salario_total}")
    except ValueError:
        print("Error: Ingrese un número válido.")

def calcular_consumo_agua():
    try:
        m3_consumidos = float(input("Ingresa los M3 de agua consumidos: "))

        if m3_consumidos < 0:
            raise ValueError("El consumo no puede ser negativo.")

        if m3_consumidos <= 4:
            subtotal = 50
        elif m3_consumidos <= 15:
            subtotal = 50 + (m3_consumidos - 4) * 8
        elif m3_consumidos <= 50:
            subtotal = 50 + (11 * 8) + (m3_consumidos - 15) * 10
        else:
            subtotal = 50 + (11 * 8) + (35 * 10) + (m3_consumidos - 50) * 11

        iva = subtotal * 0.16
        total_a_pagar = subtotal + iva

        print(f"Subtotal: {subtotal:.2f}")
        print(f"IVA (16%): {iva:.2f}")
        print(f"Total a pagar: {total_a_pagar:.2f}")
    except ValueError:
        print("Error: Ingrese un número válido.")

def calcular_promedio_final():
    try:
        calificaciones = []
        for i in range(5):
            calificacion = float(input(f"Ingresa la calificación del examen {i+1}: "))
            calificaciones.append(calificacion)

        calificacion_menor = min(calificaciones)
        calificaciones.remove(calificacion_menor)

        promedio_final = sum(calificaciones) / len(calificaciones)

        print(f"Promedio final: {promedio_final:.2f}")
        print(f"La calificación más baja eliminada: {calificacion_menor:.2f}")
    except ValueError:
        print("Error: Ingrese un número válido.")

def jugar_chinchampu():
    opciones = ["Piedra", "Papel", "Tijera"]
    try:
        jugador = input("Elige: Piedra, Papel, o Tijera: ").capitalize()
        if jugador not in opciones:
            raise ValueError("Opción no válida.")

        computadora = random.choice(opciones)

        print(f"Computadora elige: {computadora}")

        if jugador == computadora:
            print("Empate")
        elif (jugador == "Piedra" and computadora == "Tijera") or \
             (jugador == "Papel" and computadora == "Piedra") or \
             (jugador == "Tijera" and computadora == "Papel"):
            print("Ganaste")
        else:
            print("Perdiste")
    except ValueError as e:
        print(e)

def encontrar_menor_mayor():
    try:
        numeros = []
        for i in range(4):
            numero = int(input(f"Ingresa el número {i+1}: "))
            numeros.append(numero)

        menor = min(numeros)
        mayor = max(numeros)

        print(f"El menor es: {menor}")
        print(f"El mayor es: {mayor}")
    except ValueError:
        print("Error: Ingrese un número entero válido.")

def calcular_area_triangulo():
    try:
        base = float(input("Ingresa la base del triángulo: "))
        altura = float(input("Ingresa la altura del triángulo: "))

        area = (base * altura) / 2

        print(f"El área del triángulo es: {area:.2f}")
    except ValueError:
        print("Error: Ingrese un número real válido.")

def calcular_area_circulo():
    try:
        radio = float(input("Ingresa el radio del círculo: "))

        area = math.pi * radio ** 2

        print(f"El área del círculo es: {area:.2f}")
    except ValueError:
        print("Error: Ingrese un número real válido.")

def main():
    while True:
        print("\nMenu de opciones")
        print("1. Calcular promedio de calificaciones")
        print("2. Calcular salario semanal")
        print("3. Calcular consumo de agua")
        print("4. Calcular promedio final de exámenes")
        print("5. Jugar CHINCHAMPU")
        print("6. Encontrar menor y mayor de 4 números")
        print("7. Calcular área de un triángulo")
        print("8. Calcular área de un círculo")
        print("9. Salir")

        try:
            opcion = int(input("Ingrese el número de su elección: "))

            if opcion == 1:
                calcular_promedio()
            elif opcion == 2:
                calcular_salario()
            elif opcion == 3:
                calcular_consumo_agua()
            elif opcion == 4:
                calcular_promedio_final()
            elif opcion == 5:
                jugar_chinchampu()
            elif opcion == 6:
                encontrar_menor_mayor()
            elif opcion == 7:
                calcular_area_triangulo()
            elif opcion == 8:
                calcular_area_circulo()
            elif opcion == 9:
                print("Saliendo del programa...")
                break
            else:
                print("Opción no válida. Inténtalo de nuevo.")
        except ValueError:
            print("Error: Ingrese un número válido.")


main()
```
