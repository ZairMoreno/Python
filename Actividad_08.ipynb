```python
# Actividad 08 Python Zair Orlando Moreno Camarillo

import random

# Validar número entero
def valida_entero(mensaje, min_val, max_val):
    while True:
        try:
            num = int(input(mensaje))
            if min_val <= num <= max_val:
                return num
            else:
                print(f"Por favor, ingresa un número entre {min_val} y {max_val}.")
        except ValueError:
            print("Error: Ingresa un número válido.")

# Ejercicio 1
def juego_adivina_numero():
    ganados = 0
    perdidos = 0

    while True:
        numero_secreto = random.randint(1, 10)
        intentos = 3
        adivinado = False

        print("He generado un número entre 1 y 10. ¡Adivina cuál es!")

        while intentos > 0:
            intento = valida_entero("¿Cuál es tu número?: ", 1, 10)
            if intento == numero_secreto:
                print("¡Felicidades! Adivinaste el número.")
                adivinado = True
                ganados += 1
                break
            else:
                intentos -= 1
                if intentos > 0:
                    print(f"Fallaste. Te quedan {intentos} intentos.")

        if not adivinado:
            print(f"Lo siento, no adivinaste el número. Era {numero_secreto}.")
            perdidos += 1

        jugar_de_nuevo = input("¿Quieres jugar de nuevo? (s/n): ").lower()
        if jugar_de_nuevo != 's':
            break

    print(f"\nJuegos ganados: {ganados}")
    print(f"Juegos perdidos: {perdidos}")

# Generar lista de 10 números aleatorios sin repetir
def generar_lista_sin_repetidos():
    lista = []
    while len(lista) < 10:
        num = random.randint(1, 10)
        if num not in lista:
            lista.append(num)
    return lista

# Usar la función random.sample para generar la lista
def generar_lista_con_sample():
    return random.sample(range(1, 11), 10)

# Ejercicio 2
def juego_busca_numero(version):
    ganados = 0
    perdidos = 0

    while True:
        if version == 'A':
            lista = generar_lista_sin_repetidos()
        elif version == 'B' or version == 'C':
            lista = generar_lista_con_sample()

        numero_a_adivinar = random.choice(lista)
        intentos = 3
        adivinado = False

        print("He generado una lista con 10 números únicos entre 1 y 10.")
        print("Intenta adivinar en qué índice se encuentra el número generado.")

        while intentos > 0:
            indice = valida_entero(f"El número {numero_a_adivinar} se encuentra en el índice: ", 0, 9)
            if lista[indice] == numero_a_adivinar:
                print("¡Felicidades! Adivinaste el índice.")
                adivinado = True
                ganados += 1
                break
            else:
                intentos -= 1
                if intentos > 0:
                    print(f"Fallaste. Te quedan {intentos} intentos.")

        if not adivinado:
            print(f"Lo siento, no adivinaste el índice. El número {numero_a_adivinar} estaba en el índice {lista.index(numero_a_adivinar)}.")
            perdidos += 1

        jugar_de_nuevo = input("¿Quieres jugar de nuevo? (s/n): ").lower()
        if jugar_de_nuevo != 's':
            break

    print(f"\nJuegos ganados: {ganados}")
    print(f"Juegos perdidos: {perdidos}")

# Menú principal
def menu():
    while True:
        print("\nSeleccione la función que desea ejecutar:")
        print("1. Juego: Adivina el número")
        print("2. Juego: Busca el número (Versión A)")
        print("3. Juego: Busca el número (Versión B)")
        print("4. Juego: Busca el número (Versión C)")
        print("5. Salir")

        opcion = valida_entero("Ingrese el número de su elección: ", 1, 5)

        if opcion == 1:
            juego_adivina_numero()
        elif opcion == 2:
            juego_busca_numero('A')
        elif opcion == 3:
            juego_busca_numero('B')
        elif opcion == 4:
            juego_busca_numero('C')
        elif opcion == 5:
            print("Saliendo del programa...")
            break


menu()
```
