```python
#ACTIVIDAD 06 CICLOS Y FUNCIONES

print("Fuciones del programa")
print("Opcion 1. Este programa te permite ingresar n cantidad de numeros y decirte cual es el resultado de la suma, media y valor de los numeros mayores y menores.")
print("Opcion 2. Este programa genera 15 numeros impares y te despliega la media de los pares y media de impares.")
print("Opcion 3. Este programa te permite ingresar un rango las veces que desees y te despliega la cantidad de numeros y promedio de los numeros.")
print("Opcion 4. Valor que te da el area de un triangulo al ingresar los valores del mismo.")
print("Opcion 5. Programa que te regresa el promedio obtenido y dice si reprobaste, pasaste o si ya no tienes oportunidades.")

numero = int(input("Ingresa la opcion que deseas probar"))

def opcion1():
    numero = []
    while True:
        try:
            num = float(input("Ingrese el valor o pulse 0 para terminar: "))
            if num != 0:
                numero.append(num)
            else:
                break

        except ValueError:
            print("Error")
            continue

    suma = sum(numero)
    minimo = min(numero)
    maximo = max(numero)
    media = sum(numero) / len(numero)
    print("La suma de los numeros", suma)
    print("El numero mas bajo", minimo)
    print("El numero mas alto", maximo)
    print("La media de los numeros", media)


def opcion2():
    numeros_impares = []
    numeros_pares = []

    contador_impares = 0
    contador_pares = 0

    for i in range(11, 61, 2):
        if contador_impares < 15:
            numeros_impares.append(i)
            contador_impares += 1

    for i in range(10, 61, 2):
        if contador_pares < 25:
            numeros_pares.append(i)
            contador_pares += 1

    media_impares = sum(numeros_impares) / len(numeros_impares) if len(numeros_impares) > 0 else 0
    media_pares = sum(numeros_pares) / len(numeros_pares) if len(numeros_pares) > 0 else 0

    print("Números impares generados:", numeros_impares)
    print("Números pares generados:", numeros_pares)
    print("Media de los números impares:", media_impares)
    print("Media de los números pares:", media_pares)

def opcion3():
    numeros = []
    while True:
        try:
            num = float(input("Ingrese un número o pulse 0 para terminar: "))
            if num != 0:
                numeros.append(num)
            else:
                break
        except ValueError:
            print("Error: Ingrese un número válido.")

    return numeros

    while True:
        numeros = opcion3()
        cantidad_numeros = len(numeros)
        if cantidad_numeros > 0:
            promedio = sum(numeros) / cantidad_numeros
            print(f"Cantidad de números ingresados: {cantidad_numeros}")
            print(f"Promedio de los números ingresados: {promedio}")
            break
        else:
            print("No se ingresaron números.")



def opcion4():
    base = float(input("Ingrese el valor de la base del triangulo"))
    altura = float(input("Ingrese el valor de la altura del triangulo"))

    area = (base * altura) / 2
    print("El area del triangulo es: ", area)

def opcion5():
    calificaciones = []
    for i in range(3):
        calificacion = float(input("Ingrese la calificacion " + str(i+1) + " "))
        calificaciones.append(calificacion)

    print("Las calificacioes son las siguientes: ", calificaciones)

    suma = sum(calificaciones)
    promedio = suma / 3

    print("Tu promedio es: ", promedio)

    if promedio >= 60 and promedio <= 100:
        print("Felicidades avanzas al siguiente semestre")
    elif promedio < 60 and promedio >= 0:
        print("Estas reprobado")
        reprobaciones = int(input("Contando tambien esta ocacion. Cuantas veces haz repetido la materia? "))
        if reprobaciones == 3:
            print("Lastima, estas fuera de la UABC")
        elif reprobaciones == 2:
            print("Todabia puedes cursar la materia otravez, pero solo tienes una oportunidad mas")
        elif reprobaciones == 1:
            print("Recuerda que solo tienes 3 oportunidades para aprobar una materia. Te quedan 2")
        else:
            print("Hubo un problema con el programa. Probablemente tu eres la causa del problema")

if numero == 1:
    opcion1()
elif numero == 2:
    opcion2()
elif numero == 3:
    opcion3()
elif numero == 4:
    opcion4()
else:
    opcion5()
```
